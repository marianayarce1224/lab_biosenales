```python
import pandas as pd

emg = pd.read_csv('/content/emg4.csv')
ekg = pd.read_csv('/content/ekg.csv')
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 1, figsize=(12, 12))
plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, wspace=0.2, hspace=0.4)

emg_short = emg[41:]
axs[0].plot(emg_short['Unnamed: 0'], emg_short['0'])
axs[0].set_title('Señal EMG registrada', fontdict={'fontweight': 'bold', 'fontsize': 15})
axs[0].set_xlabel('Tiempo', fontdict={'fontsize': 14}, labelpad=10)
axs[0].set_ylabel('Amplitud', fontdict={'fontsize': 14}, labelpad=10)

ekg_short = ekg[30:]
axs[1].plot(ekg_short['Unnamed: 0'], ekg_short['0'])
axs[1].set_title('Señal ECG registrada', fontdict={'fontweight': 'bold', 'fontsize': 15})
axs[1].set_xlabel('Tiempo', fontdict={'fontsize': 14}, labelpad=10)
axs[1].set_ylabel('Amplitud', fontdict={'fontsize': 14}, labelpad=10)

plt.show()
```
