```python
import numpy as np
import matplotlib.pyplot as plt


k = 18
fs = 200  # Frecuencia de muestreo
T = 10  # Duracin 10segundos
t = np.arange(0, T, 1/fs)  # Vector de tiempo

# Seal
x_t = 4 * np.cos(80 * np.pi * t + np.pi/4) + 18 * np.sin(40 * np.pi * t) + 5

# Graficar
plt.figure(figsize=(30, 4))
plt.plot(t, x_t)
plt.title("Seal muestreada de $x(t)$")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.grid()
plt.show()
```
