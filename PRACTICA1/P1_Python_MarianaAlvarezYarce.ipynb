Escribir una funci√≥n que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con la nota m√≠nima, la m√°xima, media y la desviaci√≥n t√≠pica.

```python
import numpy as np
import pandas as pd

def funcion_notas(diccionario):
    # Extraer las notas del diccionario
    notas = np.array(list(diccionario.values()))
    
    # Calcular las estad√≠sticas
    nota_minima = np.min(notas)
    nota_maxima = np.max(notas)
    nota_media = np.mean(notas)
    desviacion_tipica = np.std(notas)
    
    # Crear una Serie de pandas con los resultados
    resultados = pd.Series({
        'Nota M√≠nima': nota_minima,
        'Nota M√°xima': nota_maxima,
        'Nota Media': nota_media,
        'Desviaci√≥n T√≠pica': desviacion_tipica
    })
    
    return resultados

# Ejemplo de uso
diccionario_notas = {
    'Juanita': 85,
    'Pepito': 92,
    'Mariana': 78,
    'Andrea': 88,
    'Luis': 91
}

# Obtener las estad√≠sticas
resultados = funcion_notas(diccionario_notas)
print(resultados)
```
