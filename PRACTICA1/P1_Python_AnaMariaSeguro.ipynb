<b>b)</b> Implemente la multiplicación escalar de $a\cdot b$ ¿Qué se debe realizar para poder multiplicar los dos
vectores?

Para realizar la multiplicación escalar entre vectores (producto punto), es necesario verificar que los vectores tengan las mismas dimensiones o una cantidad de elementos compatible entre sí. Para multiplicar escalarmente los vectores, se multiplican los elementos en la misma posición de ambos vectores y se suman con los productos resultantes de los demás elementos. A continuación se ilustra el proceso:
><br>Sean $a=[a_0,a_1,a_2,a_3,a_4], b=[b_0,b_1,b_2,b_3,b_4]$
<br><br>$a\cdot b=a_0*b_0+a_1*b_1+a_2+b_2+a_3*b_3+a_4*b_4$

```python
scalar_mult = np.dot(a, b)
print(scalar_mult)
```

<b>f)</b> Consulte la función de los siguientes comandos de la librería numpy: <i>ones, round, ceil, floor</i>. Implemente cada función para ejemplificar su uso


><b>numpy.ones:</b> construye un arreglo en el que todos los elementos tiene el valor de 1 (de tipo flotante por defecto). Las dimensiones del arreglo se especifican en el primer argumento y puede definirse como la cantidad de elementos o el número de filas y columnas.

><b>numpy.round:</b> redondea un número flotante a un valor entero o un valor flotante si se especifica el número de decimales con el argumento <i>decimals</i>. Puede ser usado para valores numéricos individuales y arreglos de números, para los cuales retorna el arreglo con cada elemento redondeado.

><b>numpy.ceil:</b> redondea los elementos numéricos flotantes de un arreglo al entero mayor más cercano.

><b>numpy.floor:</b> redondea los elementos numéricos flotantes de un arreglo al entero menor más cercano.

```python
ex_ones = np.ones(5)
ex_ones2 = np.ones((2, 3))
print(f'Ejemplos funcion ones:\n{ex_ones}\n{ex_ones2}\n')

ex_round = np.round(3.14159)
ex_round2 = np.round([np.e, np.pi, 9.99], decimals=2)
print(f'Ejemplos funcion round:\n{ex_round}\n{ex_round2}\n')

ex_ceil = np.ceil(3.14159)
ex_ceil2 = np.ceil([np.e, np.pi, 9.99])
print(f'Ejemplos funcion ceil:\n{ex_ceil}\n{ex_ceil2}\n')

ex_floor = np.floor(3.14159)
ex_floor2 = np.floor([np.e, np.pi, 9.99])
print(f'Ejemplos funcion ceil:\n{ex_floor}\n{ex_floor2}')
```

<h4>Escribir una función que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con la nota mínima, la máxima, media y la desviación típica.</h4>



```python
def grades_series(dic_grades):
  import pandas as pd
  grades = pd.Series(dic_grades)
  return pd.Series([grades.min(), grades.max(), grades.mean(), grades.std()], index=['min', 'max', 'mean', 'std'])

students = {'Juan': 9, 'Maria': 6.5, 'Pedro': 4, 'Carmen': 8.5, 'Luis': 5, 'Ana': 2, 'Carlos': 10, 'Sofia': 3.8}
grades_students = grades_series(students)
print(grades_students)
```
